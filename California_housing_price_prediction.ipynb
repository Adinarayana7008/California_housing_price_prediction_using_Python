**5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.**

```python
#importing numpy and pandas libraries
import numpy as np
import pandas as pd

#reading dataset and storing in a file varaible
file = pd.read_csv('housing.csv')

#dataset is converting into DataFrame and storing in a df varaible
df = pd.DataFrame(file)

#calculating mean value for total_bedrooms column from dataset
#and rounding result to 4 digits
cal_mean = round(np.mean(df.total_bedrooms),4)

#replacing NaN spaced cells with calculated mean value in total_bedrooms column
df['total_bedrooms']=df['total_bedrooms'].replace(np.NaN, cal_mean)

#calling updated dataframe
df
```

**Observations**

1) Calculated mean by using numpy library for total_bedrooms variable.

2) using replace function, replacing null value rows from total_bedrooms column to mean value.

3) Defination: The replace() method returns a copy of the string where the old substring is replaced with the new substring.

4) total_bedrooms column with out null values we are having 20433 rows × 10 columns.

5) After replacing null value rows with mean value to total_bedrooms column the data set contains 20640 rows × 10 columns.

**10. Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. If the size is 11 or more but 1000 less, it should be medium, otherwise it should be large.**

```python
# importing numpy and pandas libraries
import numpy as np
import pandas as pd

#storing total_bedrooms column values in dff varaible from dataset
dff = pd.read_csv("housing.csv")

col_list = dff["total_bedrooms"]

#empty list for strong new column values
new_list=[]

#reading list values and doing comparision and results are appending into new_list
for value in col_list:
  if value <= 10:
    new_list.append("small")
  elif value >= 11 and value < 1000:
    new_list.append("medium")
  else:
    new_list.append("large")

#Inserting the column at the 5 position after the total_bedroom column in the DataFrame
dff.insert(loc = 5,
          column = 'total_bedroom_size',
          value = new_list)

# show the dataframe
dff
```

**Observations**

1) Based on total_bedrooms column a new column is created i.e. total_bedroom_size.

2) new column is added based to total_bedrooms column

3) Now new DataFrame size is 20640 rows and 11 columns.
